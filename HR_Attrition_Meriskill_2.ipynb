# Data Visualization

```python
#Plotting a histogram for the Age distribution
plt.hist(hr_attrition['Age'], bins=10, edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
#Age count plot
sns.set(style="darkgrid")
plt.figure(figsize=(14,8))
sns.countplot(x='Age', data=hr_attrition, palette='Set1')
plt.title('Age Count Plot')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
gender_counts = hr_attrition['Gender'].value_counts()
```

```python
#Plotting a pie chart for the gender distribution
plt.figure(figsize=(8,8))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Gender Distribution')
plt.show()
```

```python
#showing the value count for employee job role
sns.set(style="darkgrid")
plt.figure(figsize=(24,18))
sns.countplot(x='Job_role', data=hr_attrition, palette='Paired')
plt.title('Job Role Plot')
plt.xlabel('Job_role')
plt.ylabel('Count')
plt.show()
```

```python
# Showing the count of employee marital status
sns.set(style="darkgrid")
plt.figure(figsize=(8,6))
sns.countplot(x='Marital_status', data=hr_attrition, palette='Set2')
plt.title('Marital Status Distribution')
plt.xlabel('Marital_status')
plt.ylabel('Count')
plt.show()
```

```python
#Which job role got paid more
plt.figure(figsize=(10,8))
sns.boxplot(data=hr_attrition, x='Monthly_income', y='Job_role')
plt.title('Who Earns More?')
plt.xticks(rotation=25)
plt.show()
```

Research Directors and Managers got paid more from the boxplot above
